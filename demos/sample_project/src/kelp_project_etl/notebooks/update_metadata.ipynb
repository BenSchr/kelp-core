```python
from databricks.sdk import WorkspaceClient
from databricks.sdk.service import jobs
w = WorkspaceClient()
dbutils = WorkspaceClient().dbutils()
```

```python
job_id = dbutils.widgets.get("job_id")
kelp_project_file = dbutils.widgets.get("kelp_project_file")
kelp_target = dbutils.widgets.get("kelp_target")
```

```python
import kelp as kp
from kelp.catalog import sync_catalog
from pyspark.sql import SparkSession

spark = SparkSession.active()

kp.init(kelp_project_file, target=kelp_target)
for query in sync_catalog():
    try:
        print(f"Executing query: {query}")
        spark.sql(query)
    except Exception as e:
        print(f"Error executing query: {query}\nException: {e}")
```

```python
w.jobs.update(
    job_id,
    new_settings=jobs.JobSettings(
        parameters=[
            jobs.JobParameterDefinition(
                name="update_metadata",
                default="false",
            )
        ],
    ),
)
```
