```python
import kelp as kp
from kelp.catalog import sync_catalog
from pyspark.sql import SparkSession

spark = SparkSession.active()

kp.init("../kelp_metadata/kelp_project.yml")
for query in sync_catalog():
    try:
        spark.sql(query)
    except Exception as e:
        print(f"Error executing query: {query}\nException: {e}")
```
